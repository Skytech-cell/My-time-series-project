```python

import pandas as pd
df=pd.read_csv('/content/HistoricalData_1734100167435.csv')
df.head()
```

```python
df=df.drop('Date',axis=1)
df.head()
```

```python
rng=pd.date_range(start="6/1/2024",end="7-30-2024",freq='B')
rng
#pd.date_range() creates a sequence of dates
#frq='B' business days(excluding weekends)
```

```python
rng=rng[:len(df)]
df.set_index(rng,inplace=True)
df
#set the index to of the dataframe
```

```python
df=df.rename(columns={'Close/Last':'Close'})
df['Close'] = df['Close'].replace('$', '', regex=True).astype(float)
#rename Close/Last into close
#remove $ dollar signs
#regex() regular expression
#astype(float) coverting string to float points
```

```python
%matplotlib inline
df.Close.plot()
```

```python
df['2024-06-03':'2024-06-11'].Close.mean()
```

```python
df.asfreq('D',method='pad')
#asfreq dataframe to change its frequency
#d data point for each calender day
#method=pad if a day is missing in the original data it will copy the value from the previous valid day to fill the gap
```

```python
rng=pd.date_range(start="1/1/2024",periods=72,freq='B')
rng
#periods=72 generate a sequence containing 72 dates
#freq='B' stands for business day(monday-friday)
```

```python
import numpy as np
ts=pd.Series(np.random.randint(1,10,len(rng)),index=rng)
ts
#generates random integer between 1(unclusive) & 10(exclusive) random integer number are dates in our time series
```
