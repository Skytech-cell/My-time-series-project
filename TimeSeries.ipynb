```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/aapl.csv',parse_dates=['Date'],index_col="Date")
data.head()
 # display the first 5 rows
 # parse_date=['date']: Date as dates converting to datetime objects
 #index_col='Date':index of the DataFrame filter based on dates
```

```python
data.index
```

```python
data.loc['2024-11']
#print specific located r & c
```

```python
data.loc['2024-11-29']
#print particular located date
```

```python
data=data.sort_index()
# sort the dataframe by index
```

```python
data.loc['2024-11-13':'2024-11-17']
#print specifies range of dates
```

```python
data=data.rename(columns={'Close/Last':'Close'})
#change data variable using rename
```

```python
data['2024-11-27':'2024-11-20']
#print specifies range of dates
```

```python
data.Close.resample('M')
#resample()= used to change the frequency of the timeSeries data
#("M")=end of the month
```

```python
data['Close']=data['Close'].str.replace('$','').str.split('$').str[0].astype(float)
#remove the dollar signs($) in the close column with empty string
#str.split('$') further splits the string
#str[0]this select the first element
#astype(float) convert into string to float data type

```

```python
data.Close.mean()
#display calculates the average of 'Close' column
```

```python
%matplotlib inline
data.Close.resample('M').mean().plot(kind='line')
```
