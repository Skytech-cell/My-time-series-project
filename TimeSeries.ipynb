```python
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday
# Import AbstractHolidayCalendar and Holiday

class Mybirthdaycalender(AbstractHolidayCalendar):
    rules = [
        Holiday('"akash"s birth day', month=12, day=17) #, observance=nearest_workday),
    ]
myc=CustomBusinessDay(calendar=Mybirthdaycalender())
myc
```
